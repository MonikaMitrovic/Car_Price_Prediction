Constructing the linear regression model

```python
# A function to compute the dot product of two vectors

def dot(xi, w):  
    n = len(xi)
    res = 0.0

    for j in range(n):
        res = res + xi[j]*w[j]
    return res 
```

```python
# example
w0 = 7.17 
w = [0.01,0.04,0.002]
```

```python
# we put everything that was built above into a single function

def train_lin_regression(X, y):

    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    
    if np.linalg.det(XTX) == 0:
        raise ValueError("The matrix X^T * X is singular and not invertible.")

    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]
```

```python
def prepare_X(df):
    df = df.copy() 
    features = base.copy()

    df['age'] = 2017 - df.year
    features.append('age')
    
    for v in [2,3,4]:
        df['num_doors_%s' % v] = (df.number_of_doors == v).astype('int')  
        features.append('num_doors_%s' % v)
    
    for c, values in categories.items():  
        for v in values:  
            df['%s_%s' % (c, v)] = (df[c] == v).astype('int')  
            features.append('%s_%s' % (c, v))

    df_num = df[features]
    df_num = df_num.fillna(0)
    X = df_num.values

    return X 
```

```python
X_train = prepare_X(df_train)
w0, w = train_lin_regression(X_train, y_train)

X_val = prepare_X(df_val) 
y_val_pred = w0 + X_val.dot(w) 

rmse(y_val, y_val_pred)
```

Regularization

```python

def train_lin_regression_regul(X, y, r = 0.01):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX = XTX + r * np.eye(XTX.shape[0])   

    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)
    
    return w_full[0], w_full[1:]
```

```python
X_train = prepare_X(df_train)
w0, w = train_lin_regression_regul(X_train, y_train, r = 0.01)

X_val = prepare_X(df_val) 
y_pred = w0 + X_val.dot(w) 

rmse(y_val, y_pred)
```

Tunning the Model

```python
for r in [0.0, 0.00001, 0.0001, 0.001, 0.1, 1, 10]:
    X_train = prepare_X(df_train)
    w0, w = train_lin_regression_regul(X_train, y_train, r = r)

    X_val = prepare_X(df_val) 
    y_pred = w0 + X_val.dot(w) 
    
    score = rmse(y_val, y_pred)
    print(f"{r:<10.5f}{w0:<40.9f}{score:<10.9f}")
```
